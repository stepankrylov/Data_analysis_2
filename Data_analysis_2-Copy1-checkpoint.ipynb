```python
list_data_frame = []
item = '1950'
names = pd.read_csv(f'names\yob{item}.txt', names=['Name', 'Gender', 'Count'])
names['Year'] = str(item)
list_data_frame.append(names)
df_res_3 = pd.concat(list_data_frame)
df_res_3
```

```python
df_res_3['FirstLetter'] = df_res_3.apply(lambda x: x['Name'][:1], axis=1)
R_count = df_res_3[df_res_3['FirstLetter'] == 'R'].groupby(['Name']).sum().sort_values('Count', ascending=False)[:10]
R_count
```

```python
list_name = []
for name, count in R_count.iterrows():
    list_name.append(name)
list_name
```

```python
plt.pie(R_count, labels=list_name)
```

```python
list_data_frame = []
for item in range(1900, 2001):
    names = pd.read_csv(f'names\yob{item}.txt', names=['Name', 'Gender', 'Count'])
    names['Year'] = str(item)
    list_data_frame.append(names)
df_res_4 = pd.concat(list_data_frame)
df_res_4
```

```python
Letters = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z']
df_sort = df_res_4.groupby(['Name']).sum().sort_values('Count', ascending=False)
df_sort
```

```python
df_sort.reset_index(inplace=True)
df_sort['CountLetter'] = df_sort.apply(lambda x: len(list(filter(lambda y: y.lower() in Letters, x['Name']))), axis=1)
df_sort
```

```python
plt.scatter(df_sort['CountLetter'], df_sort['Count'])
```
